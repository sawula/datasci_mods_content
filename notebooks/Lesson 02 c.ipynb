# 2 Google Trends data

Another dataset may provide some insight.  It reports relative interest in the US for the phrase `cold medicine` according to Google searches.  

```python
import pandas as pd
import altair as alt
```

```python
coldmed = pd.read_csv("https://raw.githubusercontent.com/sawula/Survey-of-Python-data-science-stack/gh-pages/coldmed.csv")
coldmed
```

The fields in the dataset are:

* **Interest** -- let the week with the greatest number of searches for `cold medicine` receive an `Interest` rating of 100. The other weeks receive ratings relative to that.  For example, a week with half that number of searches would receive an `Interest` rating of 50.  The name of this field is from Google.
* **Week** -- the week for which the data is collected expressed as a date. The name of this field is from Google.
* **WOY** -- the week of the year, counting up to 52.  I added this field for you to make it easier to compare data over years
* **Year** -- the year from the `Week` date.  Again, I added this field to make it easier for year-to-year comparisons.


Let's check out the `dtypes` and change `Week` to `datetime64`.  This changing of types is not ideal for a beginner, so I'll do it for you.

```python
coldmed.dtypes
```

```python
coldmed = coldmed.astype({'Week':'datetime64'})
coldmed.dtypes
```
